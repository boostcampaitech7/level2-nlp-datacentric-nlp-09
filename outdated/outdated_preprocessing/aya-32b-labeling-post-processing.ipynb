```python
import pandas as pd

df = pd.read_csv('../../data/preprocessed_v2/Lnoise_Augmentation_4981_aya-Lrecovered.csv')
print(df.shape)
df
```

```python
print(df.columns)
```

```python
df = df.assign(topic=df['topic'].str.split('\n')).explode('topic')
df
```

```python
# 'topic' drop na, empty string, and 'None'
df = df[df['topic'].notna()]
df = df[df['topic'] != '']
df = df[df['topic'] != 'None']
df
```

```python
# topic column에 'topic'이라는 단어가 포함된 행만 추출
df = df[df['topic'].str.contains('topic', case=False)]
df
```

```python
from collections import Counter

words = ' '.join(df['topic'].dropna()).split()
word_counts = Counter(words)

print(len(df))
# print("단어 빈도 수:", word_counts)
for x, y in word_counts.items():
    print(f'{x}: {y}')
```

```python
# mapping: 생활문화(0) , 스포츠(1) , 정치(2) , 사회(3) , IT과학(4) , 경제(5) , 세계(6)

mapping = {
    '생활': 0,
    '문화': 0,
    '연예': 0,
    '예술': 0,
    '날씨': 0,
    '엔터테인먼트': 0,
    '스포츠': 1,
    '게임': 1,
    '정치': 2,
    '사회': 3,
    '환경': 3,
    '과학': 4,
    'it': 4,
    '기술': 4,
    '경제': 5,
    '세계': 6
}

# mapping의 key가 string에 포함되면 value로 변경
# 기존 target 열을 보존하도록 수정된 코드
df['target'] = df.apply(
    lambda row: next((mapping[key] for key in mapping if key in str(row['topic']).lower()), row['target']),
    axis=1
)

# 결과 확인
print(df)
# df.to_csv('../../data/exp/relabel.csv', index=False)
```

```python
df['topic'] = df['topic'].str.replace(' ', '', regex=False)  # 공백 제거
df['topic'] = df['topic'].str.extract(r':(.+)$')  # ':' 오른쪽의 내용 추출
# df.to_csv('../../data/exp/relabel.csv', index=False)
```

```python
df['target'].value_counts()
```

```python
df = df[df['target'] != -1]
df['target'].value_counts()
df.to_csv('../../data/preprocessed_v2/Lnoise_Augmentation_4981_aya-Lrecovered_post_processed.csv', index=False)
```
