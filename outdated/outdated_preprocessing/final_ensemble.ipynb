```python
import pandas as pd
from sklearn.utils import resample

# 각 클래스별 데이터프레임 생성
dfs = [final2[final2['target'] == i] for i in final2['target'].unique()]

# 기준이 되는 클래스의 샘플 수 (target 값 5 기준)
undersample_size = min(len(final2[final2['target'] == 5]), 1153)

# 각 클래스에 대해 undersampling 적용
df_undersampled = pd.concat([
    resample(sub_df, replace=False, n_samples=undersample_size, random_state=42)
    if len(sub_df) > undersample_size else sub_df
    for sub_df in dfs
])

# 결과 저장 또는 출력
df_undersampled['target'].value_counts()
```
