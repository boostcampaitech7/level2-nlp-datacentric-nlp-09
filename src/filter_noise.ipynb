```python
import pandas as pd

raw = pd.read_csv('../data/raw/train.csv')
df = pd.read_csv('../data/preprocessed/noise2mask.csv')

# '*'/len ratio in text columns
df['noise_ratio'] = df['text'].str.count('\*') / df['text'].str.len()

# sort by noise ratio
df.sort_values(by='noise_ratio', ascending=True, inplace=True)
df_filtered = df.iloc[:1120].copy()
df_remaining = df.iloc[1120:].copy()

goldenid = df_filtered['ID'].to_list()
remainingid = df_remaining['ID'].to_list()


golden_data = raw[raw['ID'].isin(goldenid)].copy()
remaining_data = raw[raw['ID'].isin(remainingid)].copy()

golden_data.to_csv('../data/preprocessed/golden.csv', index=False)
remaining_data.to_csv('../data/preprocessed/remaining.csv', index=False)
```
